```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#1-Dataset
np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A','B','C','D','E'],
    'Values': [0, 70, 30, 90, 60],
    'Sales': np.random.randint(100, 500, 5),
    'Profit': np.random.randint(10, 100, 5)
})

print(df)
```

```python
# 1. area plot
# its nothing but line plot but the area is shaded

plt.figure()
plt.fill_between(df['Category'], df['Values'], color='green', alpha= 0.25)
plt.title("Ruchika")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")    #apply seaborn style
plt.figure(figsize=(7,4))
plt.fill_between(df['Category'], df['Sales'], color='skyblue', alpha= 0.5, label='Sales')
plt.fill_between(df['Category'], df['Profit'], color='orange', alpha= 0.5, label='Profit')
plt.title("Ruchika")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#2. pie chart

plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.2f%%')
plt.title("Ruchika")
plt.show()
```

```python
#3. donut chart

plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Ruchika")
plt.show()
```

```python
#4. box plot (outliers detection)

plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Ruchika")
plt.show()
```

```python
#5. heatmap (correlation)

plt.figure()
corr = df[['Sales','Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Ruchika")
plt.show
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#select only numeric columns

corr= df[['Values','Sales','Profit']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr, annot=True)
plt.title("Ruchika")
plt.show()
```

```python
# bubble plot (3 variable chart)

import matplotlib.pyplot as plt
plt.scatter(df['Sales'], df['Profit'], s=df['Values']*10)    #size of bubbles
plt.title("Ruchika")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit', size='Values', hue='Values', data=df, sizes=(50,300), palette='viridis')
plt.title("Ruchika")
plt.show()
```

```python
#2-Dataset
import numpy as np
import pandas as pd

np.random.seed(7)
n = 100
df1 = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age': np.random.randint(20, 60, n),
    'Income': np.random.randint(20000, 100000, n),
    'Loan_Amount': np.random.randint(5000, 50000, n),
    'Credit_Score': np.random.randint(300, 900, n)
})

df1['Loan_Status'] = np.where(df1['Credit_Score'] > 600, 'Approved', 'Rejected')

display(df1.head())
```

```python
# boxplot (Outlier Detection – Loan Amount)

import matplotlib.pyplot as plt

plt.boxplot(df1['Loan_Amount'])

plt.title("Ruchika")
plt.ylabel("Loan Amount")

plt.show()
```

```python
# histogram (Credit Score Distribution)

plt.hist(df1['Credit_Score'], bins=6, color='skyblue', edgecolor='black')

plt.title("Ruchika")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")

plt.show()
```

```python
# bubble Plot (Advanced Scatter)

plt.scatter(df1['Income'], df1['Loan_Amount'],
            s=df1['Credit_Score']/2,
            c=df1['Credit_Score'],
            cmap='viridis',
            alpha=0.6)

plt.colorbar(label='Credit Score')

plt.title("Ruchika")
plt.xlabel("Income")
plt.ylabel("Loan Amount")

plt.show()
```

```python
# correlation Heatmap imp viva
import seaborn as sns

corr = df1[['Age','Income','Loan_Amount','Credit_Score']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Ruchika")

plt.show()
```
